```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
```

```python
data_url = "data/gassuperpormes.csv"
data = pd.read_csv(data_url)
```

```python
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
data_cleaned = data
data_cleaned['2001'].fillna(method='ffill', inplace=True)
data_cleaned['2023'].fillna(method='ffill', inplace=True)
```

```python
# Check for missing values
print("\nMissing Values:")
print(data_cleaned.isnull().sum())
```

```python
fechas = []
precios = []

for i in range(1, 24):
    for j in range(0, 12):
        fechas.append(str(data_cleaned.columns[i]) + "-" + str(j+1).zfill(2))
        precios.append(data_cleaned.iloc[j, i])

data_transformation = pd.DataFrame({'Fecha': fechas, 'Precio': precios})

print(data_transformation)
```

```python
# plt.plot(data_transformation)
# plt.gca().set(title="Precio de combustible super por mes", xlabel="Aﾃｱo", ylabel="Precio")
# plt.show()
plt.plot(data_transformation['Fecha'], data_transformation['Precio'])
plt.gca().set(title="Precio de combustible super por aﾃｱo", xlabel="Aﾃｱo", ylabel="Precio")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
# plt.tight_layout()
plt.show()
```

```python
mediaMovil = data_transformation['Precio'].rolling(window=12).mean()
deMovil = data_transformation['Precio'].rolling(window=12).std()

# Plotting the results
plt.plot(data_transformation['Fecha'], data_transformation['Precio'], color="blue", label="Original")
plt.plot(data_transformation['Fecha'], mediaMovil, color='red', label='Media Mﾃｳvil')
plt.plot(data_transformation['Fecha'], deMovil, color='black', label='Desviaciﾃｳn Estﾃ｡ndar Mﾃｳvil')

plt.legend(loc='best')
plt.title('Media y Desviaciﾃｳn Estﾃ｡ndar Mﾃｳvil')
plt.xlabel('Aﾃｱo')
plt.ylabel('Precio')
plt.xticks(rotation=45)
# plt.tight_layout()
plt.show()
```

```python
data_transformation['Precio_Log'] = np.log(data_transformation['Precio'])
data_transformation_log = np.log(data_transformation['Precio'])
plt.plot(data_transformation.index, data_transformation['Precio_Log'])
plt.gca().set(title="Logarithmic Transformation of Precio", xlabel="Fecha", ylabel="Log(Precio)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo CSV
def read_csv_data(filepath):
    monthly_data = {}
    with open(filepath, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)
        years = [int(year) for year in header if year.isdigit()]

        for row in reader:
            month = row[0]
            values = [float(value) if value else np.nan for value in row[2:-1]]
            monthly_data[month] = values
    return years, monthly_data

def plot_by_year(year, years, monthly_data):
    """Plot data for all months for a specific year."""
    if year not in years:
        print(f"No data found for year: {year}")
        return
    
    months = list(monthly_data.keys())
    index = years.index(year)
    values = [monthly_data[month][index] for month in months]
    
    plt.figure(figsize=(10, 6))
    plt.bar(months, values, color='green')
    plt.title(f'Datos para el aﾃｱo {year}')
    plt.xlabel('Mes')
    plt.ylabel('Valor')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Path to your CSV file
filepath = 'data/gassuperpormes.csv'

# Read CSV data
years, monthly_data = read_csv_data(filepath)
def plot_average_by_month(years, monthly_data):
    """Plot average data for all months across all years."""
    months = list(monthly_data.keys())
    averages = [np.nanmean(monthly_data[month]) for month in months]
    
    plt.figure(figsize=(12, 7))
    plt.bar(months, averages, color='green')
    plt.title('Promedio por Mes a lo largo de todos los aﾃｱos')
    plt.xlabel('Mes')
    plt.ylabel('Valor Promedio')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Using the previously read data
plot_average_by_month(years, monthly_data)

```
